```python
for i in range(len(df1)):
     df1.A1[i]=2**df1.A1[i]
     df1.B1[i]=2**df1.B1[i]
     df1.C1[i]=2**df1.C1[i]
     df1.D1[i]=2**df1.D1[i]
     df1.E1[i]=2**df1.E1[i]

     df1.A2[i]=2**df1.A2[i]
     df1.B2[i]=2**df1.B2[i]
     df1.C2[i]=2**df1.C2[i]
     df1.D2[i]=2**df1.D2[i]
     df1.E2[i]=2**df1.E2[i]
```

```python
for i in range(len(df1)):
     df1.A1[i]=df1.A1[i]/(par.Ap[df1.serial[i]]/par.Ad[df1.serial[i]])
     df1.B1[i]=df1.B1[i]/(par.Bp[df1.serial[i]]/par.Bd[df1.serial[i]])
     df1.C1[i]=df1.C1[i]/(par.Cp[df1.serial[i]]/par.Cd[df1.serial[i]])
     df1.D1[i]=df1.D1[i]/(par.Dp[df1.serial[i]]/par.Dd[df1.serial[i]])
     df1.E1[i]=df1.E1[i]/(par.Ep[df1.serial[i]]/par.Ed[df1.serial[i]])

     df1.A2[i]=df1.A2[i]/(par.Ap[df1.serial[i]]/par.Ad[df1.serial[i]])
     df1.B2[i]=df1.B2[i]/(par.Bp[df1.serial[i]]/par.Bd[df1.serial[i]])
     df1.C2[i]=df1.C2[i]/(par.Cp[df1.serial[i]]/par.Cd[df1.serial[i]])
     df1.D2[i]=df1.D2[i]/(par.Dp[df1.serial[i]]/par.Dd[df1.serial[i]])
     df1.E2[i]=df1.E2[i]/(par.Ep[df1.serial[i]]/par.Ed[df1.serial[i]])
```

```python
for i in range(len(df1)):
     df1.A1[i]=math.log(df1.A1[i],2)
     df1.B1[i]=math.log(df1.B1[i],2)
     df1.C1[i]=math.log(df1.C1[i],2)
     df1.D1[i]=math.log(df1.D1[i],2)
     df1.E1[i]=math.log(df1.E1[i],2)

     df1.A2[i]=math.log(df1.A2[i],2)
     df1.B2[i]=math.log(df1.B2[i],2)
     df1.C2[i]=math.log(df1.C2[i],2)
     df1.D2[i]=math.log(df1.D2[i],2)
     df1.E2[i]=math.log(df1.E2[i],2)
```

```python
Am=Bm=Cm=Dm=Em=Fm=0
for i in range(len(df1)):
  Am+=df1.A1[i]+df1.A2[i]
  Bm+=df1.B1[i]+df1.B2[i]
  Cm+=df1.C1[i]+df1.C2[i]
  Dm+=df1.D1[i]+df1.D2[i]
  Em+=df1.E1[i]+df1.E2[i]
Am/=2*len(df1)
Bm/=2*len(df1)
Cm/=2*len(df1)
Dm/=2*len(df1)
Em/=2*len(df1)
```

```python
df1.A1=df1.A1.apply(lambda x: x-Am)
df1.B1=df1.B1.apply(lambda x: x-Bm)
df1.C1=df1.C1.apply(lambda x: x-Cm)
df1.D1=df1.D1.apply(lambda x: x-Dm)
df1.E1=df1.E1.apply(lambda x: x-Em)

df1.A2=df1.A2.apply(lambda x: x-Am)
df1.B2=df1.B2.apply(lambda x: x-Bm)
df1.C2=df1.C2.apply(lambda x: x-Cm)
df1.D2=df1.D2.apply(lambda x: x-Dm)
df1.E2=df1.E2.apply(lambda x: x-Em)
```
